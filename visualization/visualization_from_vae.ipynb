# 模型生成

```python
# 加载模型
device = "cuda:0"
net = VAE().to(device)
model_path = '/home/hanyuji/Results/VAE_result/vae_model_0603.pt'
net.load_state_dict(torch.load(model_path))
```

```python
# 准备存放输出
latent_list_mix=[]
latent_dict = {}  # 12个数组 [], [], [], [], [], [], [], [], [], [], [], []
recon_dict = {}  # 12个数组

# {4: [], 6: [], 8: []}
for index in test_index:
    latent_dict[index] = []
    recon_dict[index] = []
```

```python
net.eval()
for (x, y) in test_dataloader:
    x = x.float().to(device)
    z, mu, var = net.encoder(x)
    recon = net.decoder(z)
    
    z_np = z.detach().cpu().numpy()
    y_np = y.detach().cpu().numpy()
    recon_np = recon.detach().cpu().numpy()
        
    for (z_i,y_i) in zip(z_np,y_np):
        latent_dict[y_i].append(z_i)        
    for (recon_i,y_i) in zip(recon_np,y_np):
        recon_dict[y_i].append(recon_i)
       
```

```python
latent_list = []
recon_list = []
label_list = []
for index, arr in latent_dict.items():
    latent_list.append(np.asarray(arr))
    for _ in arr:
        label_list.append(index)
    
for index, arr in recon_dict.items():
    recon_list.append(np.asarray(arr))
```

```python
for i,j in zip(data_list,latent_list):
    print(f'{i.shape}---{j.shape}')
```

# 准备画图数据

```python
labels_list_true = []
for index,arr in enumerate(data_list):
    for _ in arr:
        labels_list_true.append(index)


true_array = np.concatenate(data_list, axis=0)
latent_array = np.concatenate(latent_list, axis=0)
recon_array = np.concatenate(recon_list, axis=0)

labels_array_true = np.asarray(labels_list_true)
labels_array_recon = np.asarray(label_list)


print('true_array:',true_array.shape)
print('labels_array_true:', labels_array_true.shape)
print('latent_array:', latent_array.shape)
print('recon_array:', recon_array.shape)
print('labels_array_recon:', labels_array_recon.shape)
```

# 可视化

```python
device = cp.cuda.Device(0)
device.use()

def linearSegmentCMap(num_colors, cmap_name="plasma"):
    '''Construct colormap for linearly segmented colors.'''
    cm = plt.get_cmap(cmap_name)
    color_list = [cm(i//3*3.0/num_colors) for i in range(num_colors)]
    return color_list
```

```python
pca_pcs = 10
n_neighbors = 50
min_dist = 0.2


# 构建PCA和UMAP模型
pca_model = PCA(n_components=pca_pcs, svd_solver="auto")
umap_model = UMAP(n_components=2, n_neighbors=n_neighbors, min_dist=min_dist)

umap_true_data = umap_model.fit_transform(pca_model.fit_transform(true_array))
# umap_recon_data = umap_model.transform(pca_model.transform(true_array))
# umap_recon_data = umap_model.fit_transform(pca_model.fit_transform(true_array))
umap_recon_data = umap_model.fit_transform(pca_model.fit_transform(recon_array))

```

```python
# 画图
'''Plot predictions at all timepoints.'''
unique_tps_true = np.unique(labels_array_true).astype(int).tolist()
unique_tps_pred = np.unique(labels_array_recon).astype(int).tolist()

n_tps = unique_tps_true[-1]+1
color_list = linearSegmentCMap(n_tps, cmap_name="plasma")
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Ground Truth", fontsize=15)
ax2.set_title("Pred Data", fontsize=15)

for t in unique_tps_true:
    
    # if t != 10:
    #     continue
    
    
    true_t_idx = np.where(labels_array_true == t)
    ax1.scatter(
        umap_true_data[true_t_idx, 0],
        umap_true_data[true_t_idx, 1],
        label=t,
        color=color_list[t],
        s=5,
        alpha=0.5,
    )
for t in unique_tps_pred:
# for t in unique_tps_true:

    true_t_idx = np.where(labels_array_recon == t)
    # true_t_idx = np.where(labels_array_true == t)
    
    ax2.scatter(
        umap_recon_data[true_t_idx, 0],
        umap_recon_data[true_t_idx, 1],
        label=t,
        color=color_list[t],
        s=5,
        alpha=0.5,
    )
ax1.legend(loc="center left", bbox_to_anchor=(1.0, 0.5))
ax2.legend(loc="center left", bbox_to_anchor=(1.0, 0.5))
plt.show()
```
